```python
import os
import tokenize

def count_lines_and_tokens(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        lines = f.readlines()
        num_lines = len(lines)
    
    # 计算token数
    with open(file_path, 'rb') as f:
        tokens = list(tokenize.tokenize(f.readline))
        num_tokens = len(tokens)
        
    return num_lines, num_tokens

total_lines = 0
total_tokens = 0
file_stats = {}

# 遍历data目录下的所有.py文件
for root, dirs, files in os.walk('data'):
    for file in files:
        if file.endswith('.py'):
            file_path = os.path.join(root, file)
            lines, tokens = count_lines_and_tokens(file_path)
            file_stats[file_path] = (lines, tokens)
            total_lines += lines
            total_tokens += tokens

# 打印每个文件的统计信息
print("每个Python文件的统计信息:")
for file_path, (lines, tokens) in file_stats.items():
    print(f"{file_path}: {lines}行, {tokens}个token")

print(f"\n总行数: {total_lines}行")
print(f"总token数: {total_tokens}个")
```

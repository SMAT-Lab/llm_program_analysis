```python
import os
import shutil

def copy_small_files():
    # 源目录列表
    src_dirs = ['../python', '../cangjie', '../ts']
    
    # 遍历每个源目录
    for src_dir in src_dirs:
        # 确保源目录存在
        if not os.path.exists(src_dir):
            print(f"目录不存在: {src_dir}")
            continue
            
        # 遍历目录中的所有文件
        for root, dirs, files in os.walk(src_dir):
            for file in files:
                src_path = os.path.join(root, file)
                
                # 读取文件行数
                try:
                    with open(src_path, 'r', encoding='utf-8') as f:
                        line_count = sum(1 for line in f)
                        
                    # 如果文件少于100行,复制到当前目录    
                    if line_count < 100:
                        dst_path = os.path.join('.', src_dir.split('/')[-1], file)
                        shutil.copy2(src_path, dst_path)
                        print(f"已复制: {src_path} -> {dst_path}")
                except Exception as e:
                    print(f"处理文件时出错 {src_path}: {str(e)}")

# 执行复制
copy_small_files()
```

```python
import os

def remove_test_files():
    # 遍历当前目录下的所有文件
    for root, dirs, files in os.walk('.'):
        for file in files:
            file_path = os.path.join(root, file)
            
            # 尝试读取文件内容
            try:
                with open(file_path, 'r', encoding='utf-8') as f:
                    content = f.read().lower()
                    
                # 检查文件内容是否包含测试相关的关键词
                if any(keyword in content for keyword in ['test', '@test', 'testcase', '@testcase', 'assert']):
                    print(f"删除测试文件: {file_path}")
                    os.remove(file_path)
                    
            except Exception as e:
                print(f"处理文件时出错 {file_path}: {str(e)}")
                
# 执行删除测试文件
remove_test_files()

```
